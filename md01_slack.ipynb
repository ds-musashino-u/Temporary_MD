```python
import pandas as pd

df = pd.read_csv('slack.csv')
df
```

```python
df.info()
```

```python
df['日付'] = pd.to_datetime(df['日付'])
df
```

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import japanize_matplotlib

plt.figure(figsize=(16, 4))
plt.plot(df['日付'],df['日間アクティブメンバー数'])
plt.ylabel('日間アクティブメンバー数')
plt.xlabel('日付')
```

```python
df['Year'] = df['日付'].dt.year 
df['Month'] = df['日付'].dt.month 
df
```

```python
import warnings
warnings.filterwarnings('ignore')

import seaborn as sns

plt.figure(figsize=(16, 8))
sns.boxplot(
    x="Year", y="日間アクティブメンバー数",
    hue="Month", palette=sns.color_palette(n_colors=12),
    data=df
)
sns.despine(offset=10, trim=True)
```

```python
import warnings
warnings.filterwarnings('ignore')

plt.figure(figsize=(16, 8))
sns.boxplot(
    x="Year", y="投稿されたメッセージ数",
    hue="Month", palette=sns.color_palette(n_colors=12),
    data=df
)
sns.despine(offset=10, trim=True)
```

```python
#散布図とヒストグラムを併せて表示する
sns.jointplot(x='日間アクティブメンバー数', y='投稿されたメッセージ数', data=df, kind='hex')
plt.show()
```
