```python
# pandas, matplotlib, seabornをインポートしてください。
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# seabornのグラフテーマをdarkgridに設定してください。
sns.set_theme(style="darkgrid")
```

```python
# 一つ上の階層にあるdatasetsディレクトリの中のusa.csvファイルを読み込み
# 先頭の5行を表示してください。
df = pd.read_csv('../datasets/usa.csv')
df.head()
```

```python
#「Height」のヒストグラムを作成してください。
sns.histplot(data=df, x='Height')
```

```python
# 「Weight」のヒストグラムを作成し、binの数を20に設定してください。
sns.histplot(data=df, x='Weight', bins=20)
```

```python
# 「Height」と「Weight」の散布図を作成してください。
sns.scatterplot(data=df, x='Height', y='Weight')
```

```python
# 「Height」の箱ひげ図を作成してください。
sns.boxplot(data=df, y='Height')
```

```python
# 「Weight」の箱ひげ図を作成してください。
sns.boxplot(data=df, y='Weight')
```

```python
#「Height」と「Weight」のジョイントプロットを作成してください。
sns.jointplot(data=df, y='Height', x='Weight')
```

```python
# 縦軸（Y軸)のラベルを「Height」
# 横軸（X軸)のラベルを「Weight」に設定してください。
plt.xlabel('Weight')
plt.ylabel('Height')

# グリッドを表示してください。
plt.grid(True)

# 散布図を表示してください。
sns.scatterplot(x='Weight',y='Height',data=df)
```
