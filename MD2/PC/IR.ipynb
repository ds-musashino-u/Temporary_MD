```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../datasets/sample/IR2.csv",names=["time", "Status"] )
df.head()
```

```python
print(df['Status'].value_counts())
```

```python
# Statusを数値に変換
df['Status'] = df['Status'].map({'Detected': 1, 'Not Detected': 0})

print(df.head())
```

```python
df['time'] = pd.to_datetime(df['time'])

sns.set_theme(style="darkgrid")
sns.lineplot(x='time', y='Status', data=df)
plt.xlabel('time')
plt.ylabel('Status (Detected=1, Not Detected=0)')
plt.title('Status over time')
plt.show()
```

```python
sns.countplot(x='Status', data=df)
plt.xticks(ticks=[0, 1], labels=['Not Detected', 'Detected'])
plt.show()
```

```python
# カテゴリ変数の数値変換
# scikit-learnの仕組みを利用して自動的に数値変換した場合

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder

df = pd.read_csv("../datasets/sample/IR.csv",names=["time", "Status"] )
print(df['Status'].value_counts())
# Statusを数値に変換
IR_label = LabelEncoder()
df['Status'] = IR_label.fit_transform(df['Status'])
df
```
