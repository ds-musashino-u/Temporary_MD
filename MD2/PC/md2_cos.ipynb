```python
sns.set_theme(style="darkgrid")
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df_1 = pd.read_csv('GitHub_MD/MD2/datasets/acc_data/No1.csv')
df_2 = pd.read_csv('GitHub_MD/MD2/datasets/acc_data/No2.csv')
df_3 = pd.read_csv('GitHub_MD/MD2/datasets/acc_data/No3.csv')

df_1['source'] = 'No1'
df_2['source'] = 'No2'
df_3['source'] = 'No3'

df = pd.concat([df_1, df_2, df_3])
df
x_col = 'relative_time'
y_col = 'AccX'
hue_col = 'source'

sns.lineplot(data=df, x=x_col, y=y_col, hue=hue_col)

# x軸、y軸、タイトルに列名を表示
plt.xlabel(x_col)
plt.ylabel(y_col)
plt.show()
```

```python
df_1 = df_1.reindex(df_2.index)
df_3 = df_3.reindex(df_2.index)

print(len(df_1))
print(len(df_2))
print(len(df_3))
```

```python
print(1 - distance.cosine(df_1['AccX'],df_2['AccX']))
print(1 - distance.cosine(df_1['AccX'],df_3['AccX']))
print(1 - distance.cosine(df_2['AccX'],df_3['AccX']))
```
