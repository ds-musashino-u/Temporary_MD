```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSVファイルの読み込み
df = pd.read_csv('data1.csv')
plt.ylabel('Score')

# グラフのテーマを設定
sns.set_style("darkgrid")

# 棒グラフの作成
# width=棒グラフの幅
# errwidth・・・エラーバー＝データのばらつきを表す（95％信頼区間）
# capsize・・・エラーバーの両端で直交する線の幅
sns.barplot(data=df, width=0.3, errwidth=1, capsize=0.05)

```
