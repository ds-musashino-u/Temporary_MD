```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# グラフのテーマを設定
sns.set_style("darkgrid")

# CSVファイルを読み込む
df = pd.read_csv('data1.csv')

# 数学と英語の相関関係について散布図と回帰直線から表す
# 下図から相関がみられないことがわかる
sns.regplot(x=df['English'],y=df['Math'])

# 相関係数を求める (負の相関）-1 〜 1（正の相関） 
# 0.7〜: 強い正の相関
# 0.3〜0.7: 中程度の正の相関
# 0.0〜0.3: 弱い正の相関
# 0.0〜-0.3: 弱い負の相関
# -0.3〜-0.7: 中程度の負の相関
# -0.7〜: 強い負の相関
print(df.corr())

```
