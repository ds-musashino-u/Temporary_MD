```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# CSVファイルを読み込む
df = pd.read_csv('data1.csv')

# 数学の得点分布を描く
sns.histplot(data=df['Math'])

# x軸を0〜100の範囲に設定
plt.xlim(0, 100)

# x軸のラベルを設定
plt.xlabel('Score')
```

```python
# 英語の得点分布を描く
sns.histplot(data=df['English'])

# x軸のラベルを設定
plt.xlabel('Score')
```

```python
# グラフのテーマを設定（より見やすくするためのデザイン変更）
sns.set_style("darkgrid")

# 英語の得点分布を描く設定 
# 下のlegendで系列名を表示するためにここでlabelを設定
# bins・・・ヒストグラムの粒度
sns.histplot(data=df['English'],label='English',bins=30)

# 数学の得点分布を描く設定
sns.histplot(data=df['Math'],label='Math',bins=15)

# x軸のラベルを設定
plt.xlabel('Score')

# 系列名（Math, English）を追加する
plt.legend()
```
